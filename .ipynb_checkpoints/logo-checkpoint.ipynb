```python
# open the file and read it into a variable mario
text = open("mario.txt", "r").read()
```

```python
text
```

```python
# instantiate a word cloud object
mario_wc = WordCloud( 
    max_font_size=100,
    background_color='white',
    collocations=False,
    colormap='bone'
)

# generate the word cloud
mario_wc.generate(text)
```

```python
# display the word cloud
plt.figure(figsize=(12,12))
plt.imshow(mario_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
```

```python
# Transform your mask into a new one that will work with the function:
transformed_cap_mask = np.ndarray((cap_mask.shape[0],cap_mask.shape[1]), np.int32)

for i in range(len(cap_mask)):
    transformed_cap_mask[i] = list(map(transform_format, cap_mask[i]))
```
